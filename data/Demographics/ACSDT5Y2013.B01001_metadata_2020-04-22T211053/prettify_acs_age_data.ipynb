```python
import pandas as pd
import numpy as np
```

```python
tracts =pd.read_csv('ACSDT5Y2018.B01001_data_with_overlays_2020-04-22T211053.csv', skiprows=[0])
tracts.head(3)
```

```python
tracts = tracts.join(pd.DataFrame(tracts['Geographic Area Name'].str.split(',').tolist(),
                                   columns = ['tract','county','city']))
tracts["county"].replace({" Bronx County": "Bronx", 
                          " Queens County": "Queens", 
                          " Kings County": "Brooklyn",
                          " Richmond County": "Staten Island",
                          " New York County": "Manhattan"}, inplace = True)
tracts['tract'] = tracts['tract'].apply(lambda x : x[13:])
tracts.head(3)
```

```python
tracts["BoroCTLbl"] = tracts["county"] + " " + tracts["tract"].astype(str)
tracts.head()
```

```python
all_columns = tracts.columns.tolist()
all_columns = list(str(x) for x in all_columns)
keep_columns = ['id','Geographic Area Name','tract','county','city','BoroCTLbl']
for i in all_columns:
    if i.find('Estimate') == 0:
        keep_columns.append(i)
keep_columns
```

```python
estimate = tracts[keep_columns]
estimate.head()
```
