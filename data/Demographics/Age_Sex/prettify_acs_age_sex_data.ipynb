```python
import pandas as pd
import numpy as np
```

```python
tracts =pd.read_csv('ACSST5Y2018.S0101_data_with_overlays_2020-05-25T153603.csv') #, skiprows=[0])
tracts.head(3)
```

```python
tracts = tracts.join(pd.DataFrame(tracts['Geographic Area Name'].str.split(',').tolist(),
                                   columns = ['tract','county','city']))
tracts["county"].replace({" Bronx County": "Bronx", 
                          " Queens County": "Queens", 
                          " Kings County": "Brooklyn",
                          " Richmond County": "Staten Island",
                          " New York County": "Manhattan"}, inplace = True)
tracts['tract'] = tracts['tract'].apply(lambda x : x[13:])
tracts.head(3)
```

```python
tracts["BoroCTLbl"] = tracts["county"] + " " + tracts["tract"].astype(str)
tracts.head()
list(filter(lambda x: x.startswith("Estimate!!Total!!Total population!!"), list(tracts.columns)))
```

```python
cols_to_keep = ['city', 'tract', 'county', "BoroCTLbl", 'Total', 'Male', 'Female', 
                '00-05', '05-09', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', 
                '40-44', '45-49', '50-54', '55-59', '60-64', '65-69', '70-74', '75-79', 
                '80-84', '85-UP']

prettier_cols = {
    'Estimate!!Total!!Total population': 'Total',
    'Estimate!!Male!!Total population': 'Male',
    'Estimate!!Female!!Total population': 'Female',
    'Estimate!!Total!!Total population!!AGE!!Under 5 years': "00-05",
    'Estimate!!Total!!Total population!!AGE!!5 to 9 years': "05-09",
    'Estimate!!Total!!Total population!!AGE!!10 to 14 years': "10-14",
    'Estimate!!Total!!Total population!!AGE!!15 to 19 years': "15-19",
    'Estimate!!Total!!Total population!!AGE!!20 to 24 years': "20-24",
    'Estimate!!Total!!Total population!!AGE!!25 to 29 years': "25-29",
    'Estimate!!Total!!Total population!!AGE!!30 to 34 years': "30-34",
    'Estimate!!Total!!Total population!!AGE!!35 to 39 years': "35-39",
    'Estimate!!Total!!Total population!!AGE!!40 to 44 years': "40-44",
    'Estimate!!Total!!Total population!!AGE!!45 to 49 years': "45-49",
    'Estimate!!Total!!Total population!!AGE!!50 to 54 years': "50-54",
    'Estimate!!Total!!Total population!!AGE!!55 to 59 years': "55-59",
    'Estimate!!Total!!Total population!!AGE!!60 to 64 years': "60-64",
    'Estimate!!Total!!Total population!!AGE!!65 to 69 years': "65-69",
    'Estimate!!Total!!Total population!!AGE!!70 to 74 years': "70-74",
    'Estimate!!Total!!Total population!!AGE!!75 to 79 years': "75-79",
    'Estimate!!Total!!Total population!!AGE!!80 to 84 years': "80-84",
    'Estimate!!Total!!Total population!!AGE!!85 years and over': "85-UP"
}
```

```python
tracts.rename(columns=prettier_cols, inplace=True)
tracts.head()
```

```python
for_export = tracts[cols_to_keep]
for_export
```

```python
for_export.to_csv('tract_age_sex_acs2018.csv',index=False)
```
